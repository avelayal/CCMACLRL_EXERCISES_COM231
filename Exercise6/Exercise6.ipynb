## 1. Train a KNN Regressor

```python
score_list = {}
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
X = df[numerical_features].drop('SalePrice', axis=1)
y = df['SalePrice']

X = X.fillna(X.mean())

param_grid = {'n_neighbors': range(1, 31)}

knn_grid = GridSearchCV(KNeighborsRegressor(), param_grid, cv=5)

knn_grid.fit(X, y)

best_n_neighbors = knn_grid.best_params_['n_neighbors']
best_knn_score = knn_grid.best_score_

print(f"Best n_neighbors for KNN Regressor: {best_n_neighbors}")
print(f"Best cross-validation score for KNN Regressor: {best_knn_score}")

knn = KNeighborsRegressor(n_neighbors=best_n_neighbors)
knn.fit(X, y)

score_list['KNN'] = best_knn_score
print(f"KNN Regressor score (with best parameters): {score_list['KNN']}")
```

- Perform cross validation

```python
from sklearn.model_selection import cross_val_score

knn_scores = cross_val_score(knn, X, y, cv=5)

print(f"KNN Regressor cross-validation scores (with best parameters): {knn_scores}")
print(f"Average KNN Regressor cross-validation score (with best parameters): {knn_scores.mean()}")
```

## 2. Train a SVM Regression

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10], 'epsilon': [0.01, 0.1, 0.5]}

svr_grid = GridSearchCV(SVR(), param_grid, cv=5)

svr_grid.fit(X, y)

best_svr_params = svr_grid.best_params_
best_svr_score = svr_grid.best_score_

print(f"Best parameters for SVR Regressor: {best_svr_params}")
print(f"Best cross-validation score for SVR Regressor: {best_svr_score}")

svr = SVR(**best_svr_params)
svr.fit(X, y)

score_list['SVR'] = best_svr_score
print(f"SVM Regressor score (with best parameters): {score_list['SVR']}")
```

- Perform cross validation

```python
from sklearn.model_selection import cross_val_score

svr_scores = cross_val_score(svr, X, y, cv=5)

print(f"SVR Regressor cross-validation scores (with best parameters): {svr_scores}")
print(f"Average SVR Regressor cross-validation score (with best parameters): {svr_scores.mean()}")
```

## 3. Train a Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': [None, 5, 10, 20], 'min_samples_split': [2, 5, 10]}

dt_grid = GridSearchCV(DecisionTreeRegressor(random_state=1), param_grid, cv=5)

dt_grid.fit(X, y)

best_dt_params = dt_grid.best_params_
best_dt_score = dt_grid.best_score_

print(f"Best parameters for Decision Tree Regressor: {best_dt_params}")
print(f"Best cross-validation score for Decision Tree Regressor: {best_dt_score}")

dt = DecisionTreeRegressor(random_state=1, **best_dt_params)
dt.fit(X, y)

score_list['Decision Tree'] = best_dt_score
print(f"Decision Tree Regressor score (with best parameters): {score_list['Decision Tree']}")
```

- Perform cross validation

```python
from sklearn.model_selection import cross_val_score

dt_scores = cross_val_score(dt, X, y, cv=5)

print(f"Decision Tree Regressor cross-validation scores (with best parameters): {dt_scores}")
print(f"Average Decision Tree Regressor cross-validation score (with best parameters): {dt_scores.mean()}")
```

## 4. Train a Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators': [50, 100, 200], 'max_depth': [None, 5, 10, 20]}

rfr_grid = GridSearchCV(RandomForestRegressor(random_state=1), param_grid, cv=5)

rfr_grid.fit(X, y)

best_rfr_params = rfr_grid.best_params_
best_rfr_score = rfr_grid.best_score_

print(f"Best parameters for Random Forest Regressor: {best_rfr_params}")
print(f"Best cross-validation score for Random Forest Regressor: {best_rfr_score}")

rfr = RandomForestRegressor(random_state=1, **best_rfr_params)
rfr.fit(X, y)

score_list['Random Forest'] = best_rfr_score
print(f"Random Forest Regressor score (with best parameters): {score_list['Random Forest']}")
```

```python
from sklearn.model_selection import cross_val_score

rfr_scores = cross_val_score(rfr, X, y, cv=5)

print(f"Random Forest Regressor cross-validation scores (with best parameters): {rfr_scores}")
print(f"Average Random Forest Regressor cross-validation score (with best parameters): {rfr_scores.mean()}")
```

## 5. Compare all the performance of all regression models

```python
print("\nCross-validation Scores after Hyperparameter Tuning:")
print(f"KNN Regressor: {score_list['KNN']}")
print(f"SVR Regressor: {score_list['SVR']}")
print(f"Decision Tree Regressor: {score_list['Decision Tree']}")
print(f"Random Forest Regressor: {score_list['Random Forest']}")

best_model_name = max(score_list, key=score_list.get)
print(f"\nBest performing model: {best_model_name}")
```

## 6. Generate Submission File

Choose the model that has the best performance to generate a submission file.

```python
sample_submission_url ='https://github.com/robitussin/CCMACLRL_EXERCISES/blob/3fd7d51ffd17863598ac3f44eeefc558171a5b73/dataset/house-prices-advanced-regression-techniques/sample_submission.csv?raw=true'
sf=pd.read_csv(sample_submission_url)

test_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES/blob/3fd7d51ffd17863598ac3f44eeefc558171a5b73/dataset/house-prices-advanced-regression-techniques/test.csv?raw=true'
dt = pd.read_csv(test_url)


dt = dt.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis=1, errors='ignore')

dt['LotFrontage'] = dt['LotFrontage'].fillna(dt['LotFrontage'].mean())

dt["SaleCondition"] = [1 if each == "Normal" else 0 for each in dt["SaleCondition"]]

numerical_features_train = df.select_dtypes(include=['int64', 'float64']).columns.drop('SalePrice')
X_test = dt[numerical_features_train].copy()

X_test = X_test.fillna(X_test.mean())

best_model_name = max(score_list, key=score_list.get)

if best_model_name == 'KNN':
    best_model = knn
elif best_model_name == 'SVR':
    best_model = svr
elif best_model_name == 'Decision Tree':
    best_model = dt
elif best_model_name == 'Random Forest':
    best_model = rfr
else:
    best_model = rfr


predictions = best_model.predict(X_test)

submission_df = pd.DataFrame({
    'Id': sf['Id'],
    'SalePrice': predictions
})

submission_df.to_csv('submission_file.csv', index=False)
print(f"Submission file created using the best performing model ({best_model_name}): submission_file.csv")
```
